```python
import os

def count_words(file_path):
    word_count = {}

    with open(file_path, 'r', encoding='utf-8') as document:
        for line in document:
            words = line.strip().split()
            for word in words:
                word = word.lower() 
                if word in word_count:
                    word_count[word] += 1
                else:
                    word_count[word] = 1

    return word_count

file_path = r'.\P1_data.txt'
count_words(file_path)
```
